```python
import sys

# 1. Create a huge RANGE (0 to 1 million)
huge_range = range(1000000)
# Measure size in bytes
print(f"Size of Range object: {sys.getsizeof(huge_range)} bytes")

# 2. Create a huge LIST (0 to 1 million)
# This forces Python to actually create every number in memory
huge_list = list(range(1000000))
print(f"Size of List object:  {sys.getsizeof(huge_list)} bytes")

# QUESTION: Look at the difference in bytes!
# The Range should be tiny (like 48 bytes).
# The List should be huge (millions of bytes).
```

```python
print("--- 1. The Floor Trap (Negatives) ---")
print(f"5 // 2  = {5 // 2}")
print(f"-5 // 2 = {-5 // 2} (Did you expect -2?)")

print("\n--- 2. The Lazy Logic (Short-Circuit) ---")
# Defined a function that screams if it runs
def crash():
    print("BOOM! I crashed!")
    return True

print("Test 1: False and crash()")
# Since first is False, Python should STOP. 'crash()' should NOT run.
result = False and crash()
print("Survived Test 1.")

print("\nTest 2: True or crash()")
# Since first is True, Python should STOP. 'crash()' should NOT run.
result = True or crash()
print("Survived Test 2.")

print("\n--- 3. The Return Value Secret ---")
# Does it print True/False or the Numbers?
print(f"10 or 20  -> {10 or 20}")
print(f"0 or 20   -> {0 or 20}")
```
