```python
# Importing libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

#Loading Datasets
df = pd.read_csv("homeprices1.csv")
print(df)
print()

#Data prepration
A = (df.bedrooms.median())
m = df.drop('price',axis = 'columns')

# Fill NaN values in 'bedrooms' column with the median 'A'
m.bedrooms = m.bedrooms.fillna(A)

#Model Training
model = LinearRegression()
model.fit(m.values,df.price.values)

#Prediction
print(model.predict([[3000,30,10]]))


print()
print("Median of given data:",A)
print()
print(m)

# Plotting with plt.scatter
plt.scatter(df['area'], df['price'])
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Area vs Price')
plt.show()
```
