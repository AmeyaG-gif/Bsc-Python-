```python
# Define the function that returns the total instead of printing
def calculate_bill(price, tax):
    total = price + tax
    return total

# Get price and tax from user input
user_price = int(input("Enter the price for the returned bill: "))
user_tax = int(input("Enter the tax for the returned bill: "))

# Call the function and save the result
final_amount = calculate_bill(user_price, user_tax)

# Print the returned amount
print(f"The returned amount is: {final_amount}")
```

```python
def get_numbers():
    return 10, 20, 30   # Returns 3 things at once!

x, y, z = get_numbers()
print(x)  # 10
print(y)  # 20
print(z)  # 30

result = get_numbers()
print(result)  # Output: (10, 20, 30)
```

```python
# Define the function
def calculator(a, b):
    add = a + b
    sub = a - b
    mul = a * b
    div = a / b
    return add, sub, mul, div  # Return all four values at once

# Call the function and catch the results in 4 variables
w, x, y, z = calculator(10, 5)

# Print them separately
print("Addition:", w)
print("Subtraction:", x)
print("Multiplication:", y)
print("Division:", z)
```

```python
# Define the function
def exam_result(marks1, marks2, marks3):
    total = marks1 + marks2 + marks3
    average = total / 3
    return total, average  # Return both total and average

# Call the function and capture the results
total_marks, average_marks = exam_result(80, 90, 75)

# Print the results
print(f"Total Marks: {total_marks}")
print(f"Average: {average_marks}")
```

```python
def cart(product, price):
    print("Product is:", product)
    print("Cost is:", price)

# Both of these work perfectly!
cart(product="bangles", price=20000)
cart(price=1200, product="shirt")    # Order reversed, but still works!
```

```python
def create_id(student_name, course):
    print(f"{student_name} is studying {course}")

# First Call – Positional Arguments
create_id("Ameya", "Data Science")

# Second Call – Keyword Arguments (Reverse Order)
create_id(course="NIOS 12th", student_name="Ameya")
```

```python
def cart(product, price=40.0):    # price has a default of 40.0
    print("Product is:", product)
    print("Cost is:", price)

cart(product="pen")               # We didn't pass price, so it uses 40.0
cart(product="handbag", price=10000) # We passed price, so it overrides and uses 10000
```

```python
def order_pizza(size, crust_type="Regular"):
    print(f"Ordering a {size} pizza with {crust_type} crust.")

# First Call – Only size (default crust will be used)
order_pizza("Large")

# Second Call – Size + Custom crust
order_pizza("Medium", "Cheese Burst")
```

```python
def m(*x):         # The star tells Python: "Accept ANY number of arguments!"
    print(x)

m(10)              # Output: (10,)
m(10, 20)          # Output: (10, 20)
m(10, 20, 30)      # Output: (10, 20, 30)
```

```python
def add_all(*numbers):
    total = 0

    for num in numbers:
        total += num

    return total


# Call 1: 3 numbers
result1 = add_all(10, 20, 30)
print("Total 1:", result1)

# Call 2: 5 numbers
result2 = add_all(1, 2, 3, 4, 5)
print("Total 2:", result2)
```

```python
# Creating a lambda function
multiply = lambda x, y: x * y

# Calling the lambda function
result = multiply(5, 4)

# Printing the result
print("Result:", result)
```

```python
without_gst_cost = [100, 200, 300, 400] # [cite: 1180]
# map will apply the lambda (x + 10) to 100, then to 200, then 300...
with_gst_cost = map(lambda x: x + 10, without_gst_cost) # [cite: 1181]

print(list(with_gst_cost)) # Output: [110, 210, 310, 410]
```

```python
nums = [1, 2, 3, 4, 5]

# Using map() + lambda to double each number
doubled_nums = map(lambda x: x * 2, nums)

# Printing the final result
print(list(doubled_nums))
```

```python
items_cost = [999, 888, 1100, 1200, 1300, 777] # [cite: 1197]
gt_thousand = filter(lambda x: x > 1000, items_cost) # [cite: 1197]

print(list(gt_thousand)) # Output: [1100, 1200, 1300] [cite: 1198, 1199, 1200]
```

```python
scores = [45, 88, 91, 50, 33, 60]

# Using filter() + lambda to extract only even numbers
even_scores = filter(lambda x: x % 2 == 0, scores)

# Printing the final result
print(list(even_scores))
```

```python
from functools import reduce # Import karna zaroori hai! [cite: 1212]

items_cost = [111, 222, 333, 444] #
# reduce will do: (((111 + 222) + 333) + 444)
total_cost = reduce(lambda x, y: x + y, items_cost) #

print(total_cost) # Output: 1110 [cite: 1213, 1215]
```

```python
from functools import reduce

factors = [1, 2, 3, 4, 5]

# Using reduce() + lambda to multiply all numbers
product = reduce(lambda x, y: x * y, factors)

print(product)
```
